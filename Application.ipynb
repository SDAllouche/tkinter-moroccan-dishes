```python
import tkinter
from tkinter import PhotoImage
from tkinter import *
from tkinter import messagebox
from tkinter import filedialog
from tkinter import ttk
from tkinter.colorchooser import *
import os
from PIL import ImageTk, Image
from tkinter.filedialog import askdirectory
import mysql.connector as mysql
```

```python
window =Tk()
window.title("Moroccan dishes")
window.state("zoomed")
window.resizable(width=False,height=False)


#open\close sidebar
def switch():
    global btnState
    if btnState is True:
        navRoot.place(x=-300, y=0)
        # resetting widget colors:
        drop.config(bg="#180e46")
        window.config(bg="#180e46")
        bar.place(x=0, y=54)
        # turning button OFF:
        btnState = False
    else:
        # make window dim:
        drop.config(bg="#180e46")
        window.config(bg="#180e46")
        # created animated Navbar opening:
        bar.place(x=-300, y=0)
        closeBtn.config(bg="#951769")
        navRoot.place(x=0, y=0)
        # turing button ON:
        btnState = True
        
#connect tkinter with database 
def connection(id,language):
    connecte=mysql.connect(host="localhost",user="root",password="",database="moroccan_dishes")
    cursor=connecte.cursor()
    if language=="French":
        cursor.execute("select dish.Name_fr,description_fr.description_fr from dish inner join description_fr on dish.id=description_fr.id_dish where id='"+ str(id) +"'")
        tabl=cursor.fetchone()
        connecte.close()
        return tabl
    elif language=="English":
        cursor.execute("select dish.Name,description.description from dish inner join description on dish.id=description.id_dish where id='" + str(id) + "'")
        tabl = cursor.fetchone()
        connecte.close()
        return tabl

#------------------------search fonctions----------------------------------------------------------------

#resetting search liste 
def updt(data):
    search.delete(0, END)
    for i in data:
        search.insert(END, i)

#add every key tap to serach bar
def entry(event):
    e.delete(0, END)
    e.insert(0, search.get(ACTIVE))

    
#update search liste in every key tap
def check(event):
    a = e.get()
    if a == 0:
        data = liste1
    else:
        data = []
        for i in liste1:
            if a.lower() in i.lower():
                data.append(i)
    updt(data)
    
#find the dish and show it 
def find():
    dic_id = {"Couscous": 9, "Tajine": 21, "Harira": 10, "Zalouk": 25, "BASTILLA": 2,
              "KHOBZ": 12, "STEAMED SHEEP HEAD": 20, "SPICY SARDINES": 19, "MINT TEA": 13, "BROCHETTES": 7,
              "SNAIL SOUP": 18, "TANJIA OR TANGIA": 22, "BRIWAT": 6, "Rfissa": 15, "Trotters": 24,
              "Bisara": 4, "Tkalia": 23, "Boulfaf": 5, "Msemen": 14, "Chebakia": 8,
              "Baghrir": 1, "Sallo": 17, "Harsha": 11, "Roast Chicken": 16, "Beef With Prunes": 3}
    plat=e.get()
    if plat=="":
        response = messagebox.showinfo("Error", "Search Bar is empty ")
    else:
        ident=dic_id[plat]
        show(ident)

#-------------------------------------------------------------------------------------------------------------

#------------------------feedback fonctions---------------------------------------------------------------------

#show emoji message 
def emoji():
    response = messagebox.showinfo("Feedback", "Thank You For Rating This Application!")

#show comment message
def submit():
    e1.delete(0, END)
    response = messagebox.showinfo("Feedback", "Thank You For Leaving Comment!")
    
#-------------------------------------------------------------------------------------------------------------
    
#------------------------home fonctions---------------------------------------------------------------------

#show clicked dish 
def show(ident):
    lang=click.get()
    if lang=="Language":
        response = messagebox.showinfo("Error", "Chose language!")
    else:
        #second window to put dish's description in
        top1 = Toplevel()
        top1.title("Dishe")
        #put databadse in variable
        tabl=connection(ident,lang)
        #creat frame insidesecond window
        frame2 = LabelFrame(top1, text=tabl[0] ,font=("",18))
        #convert tuple into list
        table=tabl[1].strip().split()
        #start to put every 15 word in list
        l=[]
        k=0
        for i in range(len(table)):
            if i==0:
                l.append(table[i])
            elif i%15!=0:
                l.append(table[i])
            else:
                l.append(table[i])
                word=" ".join(l)
                l=[]
                Label(frame2,text=word,font=("",12)).grid(row=k,column=0,sticky=W,padx=(20,40),pady=(5,5))
                k+=1
        #convert liste to string
        word1 = " ".join(l)
        Label(frame2, text=word1,font=("",12)).grid(row=k+1, column=0, sticky=W,padx=(20,40),pady=(5,15))
        frame2.grid(row=0,column=0,padx=(60,80),pady=(60,200))

#we put the image and set the width and the height for dish buttons 
def resizeTk(img1,width,height):
    img1 = Image.open(img1)
    #Resize the Image using resize method
    resized_img1= img1.resize((width,height), Image.ANTIALIAS)
    img11= ImageTk.PhotoImage(resized_img1)
    return img11

#this help us to find image path
image_list = []
text_list = []
for (dirpath, dirnames, filenames) in os.walk("Dishes"):
    for x in filenames:
        if x.endswith(".png") or x.endswith(".jpg") or x.endswith(".jpeg"):
            image_list.append(os.path.join(dirpath, x))
            text_list.append(x.split('.')[0])

btnState = False
#lists of icons
liste = [PhotoImage(file="./Icons/home.png"), PhotoImage(file="./Icons/serch.png"), PhotoImage(file="./Icons/feed.png"),
         PhotoImage(file="./Icons/feed2.png"), PhotoImage(file="./Icons/submit.png"),PhotoImage(file="./Icons/logow.png")]
liste_image = [PhotoImage(file="./Icons/excelent.png"), PhotoImage(file="./Icons/vgood.png"), PhotoImage(file="./Icons/good.png"),
               PhotoImage(file="./Icons/bad.png"), PhotoImage(file="./Icons/vbad.png")]
navIcon = PhotoImage(file="./Icons/menu (2).png")

#fill list with resized images
l = [resizeTk(i, 240, 260) for i in image_list]

#-------------------------------------------------------------------------------------------------------------

#------------------------scrollbarr----------------------------------------------------------------

#creat scrollbar
vscrollbar = ttk.Scrollbar(window, orient=VERTICAL)
vscrollbar.pack(fill=Y, side=RIGHT, expand=FALSE)

#creat canvas to put scrollbar in
canvas = Canvas(window, bd=0, highlightthickness=0, yscrollcommand=vscrollbar.set, height=750, width=1180, bg="#180e46")
canvas.place(x=80, y=100)

#give scrollbar command 
vscrollbar.config(command=canvas.yview)

# create root frame inside canvas
root = Frame(canvas, bg="#180e46")
root.pack(side=LEFT, anchor=NW)
root_id = canvas.create_window(0,0, window=root, anchor=NW)

#bind root with scrollbar 
def _configure_root(event):
    size = (root.winfo_reqwidth(), root.winfo_reqheight() + 60)
    canvas.config(scrollregion="0 0 %s %s" % size)
    if root.winfo_reqwidth() != canvas.winfo_width():
        # update the canvas's width to fit the inner frame
        canvas.config(width=root.winfo_reqwidth())
root.bind('<Configure>', _configure_root)

#-------------------------------------------------------------------------------------------------------------

#------------------------Sidebar & navigator bar----------------------------------------------------------------

#creat frame for each sidebar's widget
home_frame = Frame(root, bg="#180e46")
search_frame = Frame(root, bg="#180e46")
feed_frame = Frame(root, bg="#180e46")
about_frame = Frame(root, bg="#180e46")

#creat each sidebar's widget
def creat_widget(n):
    global search_frame, home_frame, feed_frame, about_frame,liste_name
    #creat home widget
    if n == 0:
        search_frame.pack_forget();feed_frame.pack_forget();about_frame.pack_forget()
        home_frame.pack(side=LEFT, anchor=NW,padx=(30,0))
        liste_name = ["Baghrir","BASTILLA","Beef With Prunes","Bisara","Boulfaf",
                      "BRIWAT","BROCHETTES","Chebakia","Couscous","Harira",
                      "Harsha","KHOBZ","MINT TEA","Msemen","Rfissa",
                      "Roast Chicken","Sallo","SNAIL SOUP","SPICY SARDINES","STEAMED SHEEP HEAD",
                      "Tajine","TANJIA OR TANGIA","Tkalia","Trotters", "Zalouk"]
        
        #buttons with dish's names and images in it
        j = 0;
        k = 0
        for i in range(1,26):
            b1 = Button(home_frame, text=text_list[i-1],image=l[i-1], compound=TOP,command=lambda i=i:show(i)).grid(row=k, column=j, sticky=NSEW, padx=10,
                                                                          pady=(5, 0))
            j += 1
            if j == 5:
                j = 0
                k += 1
        btnState = True
        switch()
        
    #creat search widget
    elif n == 1:
        global search, liste1,listbox_frame,lbscrollbar,e,search_label
        home_frame.pack_forget();feed_frame.pack_forget();about_frame.pack_forget()
        search_frame.pack(side=LEFT, anchor=NW,padx=(200,0))
        #search label
        search_label=Label(search_frame, text="Search", font=("Monotype Corsiva", 100), fg="white", bg="#180e46")
        search_label.grid(row=0,column=0,padx=(250,0))
        #search bar
        e=Entry(search_frame, width=40, font=("arial", 14), bd=3)
        e.grid(row=1,column=0,padx=(250,0),ipady=10)
        #search button
        Button(search_frame, image=liste[5],bg="#180e46",bd=0,command=find).grid(row=1, column=1,padx=(20,0))
        #scrollbar foe listbox
        listbox_frame = Frame(search_frame)
        lbscrollbar = ttk.Scrollbar(listbox_frame, orient=VERTICAL)
        lbscrollbar.pack(fill=Y, side=RIGHT)
        #listbox to put all dishes
        search = Listbox(listbox_frame, width=71, yscrollcommand=lbscrollbar.set)
        search.pack()
        lbscrollbar.config(command=search.yview)
        liste1 = liste_name
        updt(liste1)
        search.bind("<<ListboxSelect>>", entry)
        e.bind("<KeyRelease>", check)
        listbox_frame.grid(row=2,column=0,padx=(250,0))
        btnState = True
        switch()

    #creat feedback widget
    elif n == 2:
        global e1
        home_frame.pack_forget();search_frame.pack_forget();about_frame.pack_forget()
        feed_frame.pack(side=LEFT, anchor=NW, padx=(400, 0))
        liste_name1 = ["Excelent", "Very Good", "Good", "Bad", "Very Bad"]
        #rating lable
        Label(feed_frame, text="Rate This Application !!", fg="white", bg="#180e46",font=("",11)).grid(row=0, column=0, pady=20,columnspan=5, sticky=W)
        #emoji labels and buttons
        for j in range(5):
            Label(feed_frame, text=liste_name1[j], fg="white", bg="#180e46").grid(row=1, column=j, padx=15, pady=5)
            Button(feed_frame, image=liste_image[j], bg="#180e46", bd=0, command=emoji).grid(row=2, column=j, sticky=N,padx=20, pady=20)
        #comment label
        Label(feed_frame, text="Leave a Comment !!", fg="white", bg="#180e46",font=("",11)).grid(row=3, column=0, pady=(20, 5),columnspan=5, sticky=W)
        #comment bar
        e1 = Entry(feed_frame, width=20, )
        e1.grid(row=4, column=0, columnspan=5, sticky=W, ipady=6, ipadx=150, padx=10)
        #submit button
        Button(feed_frame, image=liste[4], bg="#180e46", bd=0, command=submit).grid(row=4, column=1, columnspan=5,sticky=E)
        btnState = True
        switch()

    #creat about widget
    elif n == 3:
        home_frame.pack_forget();search_frame.pack_forget();feed_frame.pack_forget()
        about_frame.pack(side=LEFT, anchor=NW,padx=(200,0))
        liste2=["SALAHEDDINE ALLOUCHE"]
        #labels of informations
        Label(about_frame, text="Morrocain Plats", bg="#180e46", fg="white",font=("Times New Roman (Headings CS)", 36)).grid(row=0, column=0,padx=180,sticky=W,)
        Label(about_frame, text="Version : 1.0 (20/11/2021)", bg="#180e46", fg="white",font=("",11)).grid(row=1, column=0, padx=240,sticky=W,)
        frame=LabelFrame(about_frame,text="Powered By :",bg="#180e46",fg="white",font=("",14))
        for i in range(len(liste2)):
            Label(frame,text=liste2[i],bg="#180e46",fg="white").grid(row=i,column=0,padx=(30,350),sticky=W,pady=5)
        Label(about_frame, text="Copyright Â© 2021-2022 LST IRM", bg="#180e46", fg="white", font=("", 11)).grid(row=3,column=0,padx=180,sticky=W, )
        frame.grid(row=2,column=0,padx=(160,0),pady=20)
        btnState = True
        switch()


# top Navigation bar:
topFrame = Frame(window,bg="#180e46")
topFrame.pack(side="top", fill=X)

# language dropbox :
lang = ["French", "English"]
click = StringVar()
click.set("Language")
drop = OptionMenu(topFrame, click, *lang)
drop.pack(side="right", padx=20, pady=10)
drop.config(bg="#180e46", activebackground="#060213", fg="white", width=10, height=2, activeforeground="white",
            highlightthickness=0)

# Navbar button:
navbarBtn = Button(topFrame, image=navIcon, bg="#0B0424", activebackground="#951769", bd=0, command=switch)
navbarBtn.place(x=0, y=0, width=60, height=60)

# setting Navbar frame:
navRoot = Frame(window, bg="#0B0424", height=1000, width=300)
navRoot.place(x=-300, y=0)
Label(navRoot, font="Bahnschrift 15", bg="#951769", fg="white", height=2, width=300, padx=20).place(x=0, y=0)

# set y-coordinate of Navbar widgets:
y = 80
# option in the navbar:
options = ["Home", "Search", "Feedback", "About"]
# Navbar Option Buttons:
for i in range(4):
    Label(navRoot, text=options[i], bg="#0B0424", font="BahnschriftLight 15", fg="white", activebackground="#951769",
          activeforeground="green", bd=0).place(x=80, y=y + 10)
    Button(navRoot, image=liste[i], bg="#0B0424", activebackground="#951769", bd=0,
           command=lambda i=i: creat_widget(i)).place(x=20, y=y)
    y += 60

# Navbar Close Button:
closeBtn = Button(navRoot, image=navIcon, bg="#0B0424", activebackground="#0B0424", bd=0, command=switch)
closeBtn.place(x=0, y=0, width=60, height=54)

# frame side bar 
bar = Frame(window, bg="#0B0424", padx=60, width=60, height=1000)
bar.place(x=0, y=54)

# little frame for navbutton
minibar = Frame(window, bg="#951769", width=5, height=54)
minibar.place(x=0, y=0)

#---------------------------------------------------------------------------------------------------

creat_widget(0)
switch()



window.mainloop()
```
